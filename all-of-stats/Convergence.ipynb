```python
import numpy as np
import scipy.stats as scstats
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(18111842)
```

```python
n_samples = 1000
X = rng.normal(0, 1, n_samples)
Y = -rng.normal(0, 1, n_samples)
```

```python
ecdf_x = scstats.ecdf(X)
ecdf_y = scstats.ecdf(Y)
```

```python
values = np.linspace(-3, 3)
plt.plot(values, ecdf_x.cdf.evaluate(values), label='X ~ N(0, 1)')
plt.plot(values, ecdf_y.cdf.evaluate(values), label='Y ~ -N(0, 1)')
plt.legend()
plt.title('Distribution functions')
plt.plot()
```

```python
mX = np.abs(X)
values = np.linspace(0, 3)
ecdf_mx = scstats.ecdf(mX)
plt.plot(values, ecdf_mx.cdf.evaluate(values), label='X ~ |N(0, 1)|')
plt.plot(values, ecdf_x.cdf.evaluate(values), label='X ~ N(0, 1)')
plt.legend()
_ = plt.title('Distribution functions of X and |X|')
plt.plot()
```
