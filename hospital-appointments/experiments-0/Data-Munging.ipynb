This notebook contains code to bring the data in the form suitable for further analysis. We pick up from the conclusion section of the notebook 'Preliminary-analysis'.

```python
import numpy as np
import pandas as pd
import datetime
```

```python
raw_data = pd.read_csv('all_data.csv', dtype={'PatientId': str, 'AppointmentID': str})
```

```python
#raw_data['ScheduledDate'] = [datetime.date.fromisoformat(s[0:10]) for s in raw_data['ScheduledDay']]
#raw_data['ScheduledTime'] = [datetime.time.fromisoformat(s[11:19]) for s in raw_data['ScheduledDay']]
#raw_data['AppointmentDate'] = [datetime.date.fromisoformat(s[0:10]) for s in raw_data['AppointmentDay']]
raw_data['Gap'] = pd.to_datetime(raw_data['AppointmentDay']) - pd.to_datetime(raw_data['ScheduledDay'])
raw_data['Gap.d'] = raw_data['Gap'].apply(lambda x: x.days)
raw_data['PatientId'] = raw_data['PatientId'].apply(str)
raw_data['AppointmentID'] = raw_data['AppointmentID'].apply(str)
```
