We consider a one dimensional random walk in which a particle takes a step along the positive $x$-axis if a coin toss gives a head and in the opposite direction if the toss gives a tail. We are interested in exploring:

- How frequently does a particle cross a level?
- How many steps does it take to cross a level for the first time?

We do not expect an intrinsic limit beyond which a particle cannot go. Yet we expect that it will take a large number of steps before a particle goes a distance $x > 0$, for large $x$.
